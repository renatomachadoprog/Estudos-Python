Copie o loop de “Raízes quadradas”, na página 111, e encapsule-o em uma função chamada mysqrt que receba a como parâmetro, escolha um valor razoável de x e devolva uma estimativa da raiz quadrada de a.

Para testar, escreva uma função denominada test_square_root, que exibe uma tabela como esta:
a   mysqrt(a)     math.sqrt(a)  diff
-   ---------     ------------  ----
1.0 1.0           1.0           0.0
2.0 1.41421356237 1.41421356237 2.22044604925e-16
3.0 1.73205080757 1.73205080757 0.0
4.0 2.0           2.0           0.0

8.0 2.82842712475 2.82842712475 4.4408920985e-16
9.0 3.0           3.0           0.0


```python
import sys
import math
def mysqrt(a):
    x = 3
    while True:
        y=(x+a/x)/2
        if abs(y-x) < sys.float_info.epsilon:
            break
        x=y
    return x
def test_square_root():
    print('a  ', 'mysqrt(a)    ', 'math.sqrt(a) ', 'diff')
    print('-   ---------     ------------  ----')
    a = 1
    while a<10:
        print (f"{a:<3.1f} {mysqrt(a):<13.11f} {math.sqrt(a):<13.11f} {float(mysqrt(a)-math.sqrt(a))}")
        a=a+1
        
    
test_square_root()   
```

Escreva uma função chamada eval_loop que iterativamente peça uma entrada ao usuário, a avalie usando eval e exiba o resultado.

Ela deve continuar até que o usuário digite done; então deverá exibir o valor da última expressão avaliada.

```python
import math
def eval2():
    print("digite 'done' para parar!")
    string = input()
    while string !='done':
        print(eval(str(string)))
        string = input()
eval2()
```

```python


def printNexemplo(s, n):
    if n <=0:
        return
    print(s)
    printNexemplo(s, n-1)


# na página 81, usando a iteração em vez da recursividade.
def printN(s,n):
    while n!=0:
        print(s)
        printN(s,n-1)
        break
    
printN('oi',4)
```
