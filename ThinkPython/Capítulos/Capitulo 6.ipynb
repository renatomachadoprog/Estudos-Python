```python
# 6.5 mais recursividade : ainda n entendi a brisa pela recursividade

def factorial(n):
    if n == 0:
        return 1
    else: 
        recurse = factorial(n-1)
        result =n*recurse
        return result
factorial(1)
```

```python
#6.7 Mais recursividade dessa vez usando fibonacci de exempl

def fibonacci(n):
    if n == 0:
        return 0
    elif n ==1:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)
fibonacci(3)
```
