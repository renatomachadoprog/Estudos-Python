```python
#2.10 Exercicios 

#ex 2.2 -- modo interpretativo e testes -- 

#O volume de uma esfera com raio r é Fórmula – Volume de uma esfera.. Qual é o volume de uma esfera com raio 5?
r=5
pi = 3.14
volumeEsfera = 4/3 * (pi * r**3)
print('volume de uma esfera raio 5: ',volumeEsfera)

#Suponha que o preço de capa de um livro seja R$ 24,95, mas as livrarias recebem um desconto de 40%.
#O transporte custa R$ 3,00 para o primeiro exemplar e 75 centavos para cada exemplar adicional.
#Qual é o custo total de atacado para 60 cópias?
precoLivro = 24.95 - (24.95*0.4)
primeiroExemplar = precoLivro + 3
exemplarAdicional = precoLivro +0.75
totalLivros = primeiroExemplar + (exemplarAdicional*59)
print('custo de atacado para 60 cópias de um livro: R$',totalLivros)

#Se eu sair da minha casa às 6:52 e correr 1 quilômetro a um certo passo (8min15s por quilômetro),
#então 3 quilômetros a um passo mais rápido (7min12s por quilômetro) e 1 quilômetro no mesmo passo usado em primeiro lugar, 
#que horas chego em casa para o café da manhã?

umKM = 60*8+15
tresKM = 3*(60*7+12)
tempogasto = umKM*2 +tresKM
horasaida=6*3600+52*60
horadocafe = (horasaida + tempogasto) / 3600
hora = int(horadocafe)
minuto = int((horadocafe - hora) *60)

print(f"{hora}:{minuto}")


```
