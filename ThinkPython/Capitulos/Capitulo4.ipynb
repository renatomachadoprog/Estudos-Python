```python
import turtle 
import math
bob = turtle.Turtle()
def polygon(t,n,l):
    angle = 360/n
    for i in range(n):
        t.fd(l)
        t.lt(angle)

def circle(t, r):
    circumference = 2 * math.pi * r #formula 2.pi.r calcula a circunferencia
    n = 50 # numero de segmento que faz o polygon desenhar um poligono de 50 lados, que se parece c um circulo com o raio de R
    length = circumference/n
    polygon(t,n,length)


#A interface de uma função é um resumo de como ela é usada: 
#Quais são os parâmetros? O que a função faz?
#E qual é o valor de retorno?
#Uma interface é “limpa” se permitir à pessoa que a chama fazer o que quiser sem ter que lidar com detalhes desnecessários.

circle(bob,100)
```

```python
import turtle 
import math
bob = turtle.Turtle()
def polygon(t,n,l):
    angle = 360/n
    for i in range(n):
        t.fd(l)
        t.lt(angle)

def circle(t, r):
    circumference = 2 * math.pi * r #formula 2.pi.r calcula a circunferencia
    n = int(circumference / 3) +1 # numero de segmento que faz o polygon desenhar um poligono de 50 lados, que se parece c um circulo com o raio de R
    length = circumference/n
    polygon(t,n,length)
circle(bob, 50)
```

```python
import turtle 
import math
bob = turtle.Turtle()

def polyline(t,n,length,angle):
    for i in range(n):
        t.fd(length)
        t.lt(angle)

def polygon(t, n, length):
    angle = 360 / n
    polyline(t,n, length, angle)

def arc (t, r, angle):
    arcLength =2 * math.pi * r * angle/360
    n = int(arcLength /3)+1
    step_length  = arcLength / n
    step_angle  = float(angle / n)
    polyline(t, n, step_length, step_angle)
def circle(t,r):
    arc(t,r,360)

arc(bob, 50, 90)
turtle.done()
```
