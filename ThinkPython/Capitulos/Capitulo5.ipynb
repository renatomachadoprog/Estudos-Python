```python
# 5.8 Recursivdade 
# uma função que chama a si mesmo ou outra função é chamada de recursividade, visto no exemplo abaixo:

def countdown(n):
    """ 
    n : quantidade de vezes até o blastoff
    """
    if n <=0:
        print('blastoff!')
    else:
        print(n)
        countdown(n-1)

countdown(3)

def printN(s, n):
    if n <=0:
        return
    print(s)
    printN(s, n-1)
printN('oi',6)
```

```python
import time
segundostotais = time.time() 

segundosporminuto = 60
segundosporhora = 3600
segundospordia = 86400

dias_desde_a_epoca = int(time.time() / segundospordia)
segundosrestantes = segundostotais % segundospordia


horas = int(segundosrestantes // segundosporhora)
minutos = int((segundosrestantes % segundosporhora) / segundosporminuto)
segundos = int(segundosrestantes % segundosporminuto)

print(f"dias desde a epoca: {dias_desde_a_epoca}")
print(f'hora atual (utc) {horas:02d}:{minutos:02d}:{segundos:02d}')
```
