**<h3>7.3 Instrução While</h3>**

"remake" da função countdown:

```python
def countdown(n):
    while n > 0:
        print(n)
        n=n-1
    print("blastoff")
print(countdown(11))

def sequence(n):
    while n!=1:
        print(n)
        if n%2==0:
            n=n/2
        else:
            n=n*3+1
print(sequence(3))
```

**<h3>7.4  -- Break</h3>**

Break só pode ser usado dentro de LOOPs.
SUPONHA que voce quer receber uma entrada do usuário até que este digite done.

```python
while True:
    line = input('> ')  
    if line == 'done':
        break
    print(line)
print('done!')
```

**<h3>7.5 Raízes Quadradas</h3>**

Loops são usados muitas vezes em programas que calculam resultados numericos.
Começando com uma linha e os aprimorando com o tempo:

```python
a=4
x=3
y = (x+a/x)/2
print(y)
```

```python
x=y
y=(x+a/x)/2
y
```

```python
while True:
    print(x)
    y=(x+a/x)/2
    if y==x:
        break
    x=y
```

para a maior parte dos valores de 'a', funciona bem,
mas pode ser perigoso testar a igualdade de um float.
A maioria dos números racionais, como 1/3, e números irracionais como √2, não podem ser represntados exatamente com float.

e, em vez de verificar se X e Y sao iguais, é mais seguro usar a função integrada 'abs'
para calcular o valor absoluto ou a magnitude da diferenca entre eles

```python
import sys # no livro, ele usa somente 'epsilon', porém só consegui importando a lib sys
a=4
x=3
y = (x+a/x)/2
x=y
y=(x+a/x)/2
while True:
    print(x)
    y=(x+a/x)/2
    if abs(y-x) < sys.float_info.epsilon:
        break
    x=y
```

**<h3> 7.7 - Depuração </h3>**
